```python
import pandas as pd
import dblib.index as index
import numpy as np 

dire = '/run/media/alexxruzza/b98e4720-8ade-4e0d-be1f-1c6409a206f4/@home/alessandro/UNI/TESI/dust-busters/database/'

index.set_paths(pdir=dire, idir=dire)
data = pd.DataFrame(index.get_data())
data = data.set_index('KEY')

```

```python
m_data = data[(data.ALPHASS==0.005) & (data.GDRATIO==100) & (data.INCL==0) & (data.MPLANET0>2.47) & (data.MPLANET0<2.48) & (data.TIME==200)]
```

```python
m_data
```

```python
m_data.sort_values(by=['BMAJ'], inplace = True)
```

```python
from tqdm import tqdm
import matplotlib.pyplot as plt
from astropy.io import fits

import pylab as py
#import radial_data as rad

n_cols = 5
n_rows = 5
fig, axes = plt.subplots(n_rows,n_cols)
fig.set_size_inches(n_cols*6, n_rows*6)

images = []
radial_prof = []
for key, element in tqdm(m_data.iterrows()):
    X = (np.arange(1, 1024 + 1) - 513) * element['CDELT1'] *3600
    Y = (np.arange(1, 1024 + 1) - 513) * element['CDELT2'] *3600
    xx, yy = np.meshgrid(X,Y)
    
    try:
        with fits.open(f"{dire}{element['FILENAME']}") as hdu:
            images.append(hdu[0].data)
    except FileNotFoundError:
        print(f"{element['FILENAME']} not found")
        
    #radial_prof.append(rad.radial_data(images[-1][::-1], x=xx, y=yy))

```

```python
import matplotlib
from matplotlib.patches import Ellipse
matplotlib.use("pgf")

matplotlib.rcParams.update({
    "pgf.texsystem": "pdflatex",
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
})

n_cols = 5
n_rows = 2
fig, axes = plt.subplots(n_rows,n_cols)
fig.set_size_inches(n_cols*3, n_rows*3)
axes = axes.flatten()
i=0
for key, element in tqdm(m_data.iterrows()):
    im = axes[i].imshow(images[i][::-1], extent=(np.amin(X), np.amax(X), np.amin(Y), np.amax(Y)),  cmap='inferno')
    #fig.colorbar(im, ax=axes[i])
    im = axes[i].set_title(f"Beam FWHM$ = {element['BMAJ']:.2f} ''$")
    axes[i].set_xlabel("$\Delta$ RA [$''$]")
    axes[i].set_ylabel("$\Delta$ Dec [$''$]")
    
    
    
    dx, dy = (0.125, 0.125)
    beam = Ellipse(
        axes[i].transLimits.inverted().transform((dx, dy)),
        width=element['BMAJ'],
        height=element['BMAJ'],
        angle=0,
        fill=True,
        color="grey",
    )
    axes[i].add_patch(beam)
    
    i += 1

#plt.savefig('diff_mass.png')
fig.tight_layout()
plt.savefig('res_diff.pgf', dpi=400, bbox_inches='tight')
plt.show()
```
